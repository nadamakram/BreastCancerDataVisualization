# Description
Visualizing real-world dataset containing information collected from microscopic images of breast cancer tumors, similar to the image below.
<img src="https://i.imgur.com/qUESsJe.png">
Each tumor has been labeled as either **benign** (_noncancerous_) or **malignant** (_cancerous_).

## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load the data

```python
data_b_path = r"E:\DataScienceProjects\DataVisualization\cancer_b.csv"
data_m_path = r"E:\DataScienceProjects\DataVisualization\cancer_m.csv"

benign_data = pd.read_csv(data_b_path, index_col="Id")
malignant_data = pd.read_csv(data_m_path, index_col="Id")
```

```python
benign_data.head()
```

```python
malignant_data.head()
```

### Histogram
we will use histograms to show the distribution of Area(mean) for benign and malignant tumors

```python
#histogram of benign data to show the distribution of Area(mean)
sns.distplot(a = benign_data['Area (mean)'], kde=False, label="Benign Tumor")
sns.distplot(a = malignant_data['Area (mean)'], kde=False, label="Malignant Tumor")
plt.title("Distribution of Benign and Malignant based on 'Area Mean'")
#show the legend
plt.legend()
```

We can notice that Malignant Tumor have higher values of Area(mean)

### KDE plot 
we will use KDE plots to show the distribution of Radius (worst) for benign and malignant tumors

```python
sns.kdeplot(data=benign_data['Radius (worst)'], shade=True, label="Benign Tumor")
sns.kdeplot(data=malignant_data['Radius (worst)'], shade=True, label="Malignant Tumor")
plt.title("Distribution of Benign and Malignant based on 'Radius'")
```

Tumors with high radius are most likely malignant
